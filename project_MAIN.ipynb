# train_valid_test_split

```python
target = df['critical_temp']
features = df.drop(columns=['critical_temp'])
```

```python
from sklearn.model_selection import train_test_split

# dividing into training and TEMP sets
train_features, TEMP_features, train_target, TEMP_target =train_test_split(features, target, test_size=0.2, random_state=42)

# dividing TEMP into further validation and testing
valid_features, test_features, valid_target, test_target =train_test_split(TEMP_features, TEMP_target, test_size=0.5, random_state=42)

# Save the indices of the training, validation, tetsing set
train_indices = train_features.index
valid_indices = valid_features.index
test_indices = test_features.index

print("Training set shape:", train_features.shape)
print("Validation set shape:",valid_features.shape)
print("Testing set shape:",test_features.shape)
```

When you specify a value for random_state, such as random_state=42, scikit-learn uses that value as the seed for its random number generator. This ensures that the data splitting process is deterministic, meaning it will produce the same results each time you run the code with the same random_state value.

# Principal Component Analysis (PCA)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 1: Standardize the Data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 2: Compute the Covariance Matrix
cov_matrix = np.cov(scaled_features, rowvar=False)

# Step 3: Compute Eigenvectors and Eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 4: Select Principal Components
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
sorted_indices = np.argsort(explained_variance_ratio)[::-1]  # Sort eigenvalues in descending order
selected_indices = sorted_indices[:10]  # n_components is the number of principal components to keep

# Step 5: Project Data onto Principal Components
pca = PCA(n_components=10)
principal_components = pca.fit_transform(scaled_features)

# Optionally, you can access the explained variance ratio of the selected components
explained_variance_ratio_selected = pca.explained_variance_ratio_

print(cov_matrix)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, f1_score
import numpy as np

# Initialize Linear Regression model
lr_model = LinearRegression()

# Fit the model to the training data using principal components
lr_model.fit(principal_components[train_indices], train_target)  # Replace features_train with principal_components[train_indices]

# Predict on training, validation, and testing sets using principal components
train_preds = lr_model.predict(principal_components[train_indices])
valid_preds = lr_model.predict(principal_components[valid_indices])
test_preds = lr_model.predict(principal_components[test_indices])

# Calculate metrics
train_accuracy = lr_model.score(principal_components[train_indices], train_target)
valid_accuracy = lr_model.score(principal_components[valid_indices], valid_target)
test_accuracy = lr_model.score(principal_components[test_indices], test_target)
mse = mean_squared_error(test_target, test_preds)
r2 = r2_score(test_target, test_preds)

# Define threshold values for F1 score calculation
threshold = 0.5

# Binarize predictions and target values for F1 score calculation
train_preds_binary = np.where(train_preds >= threshold, 1, 0)
valid_preds_binary = np.where(valid_preds >= threshold, 1, 0)
test_preds_binary = np.where(test_preds >= threshold, 1, 0)
train_target_binary = np.where(train_target >= threshold, 1, 0)
valid_target_binary = np.where(valid_target >= threshold, 1, 0)
test_target_binary = np.where(test_target >= threshold, 1, 0)

# Calculate F1 score
f1_train = f1_score(train_target_binary, train_preds_binary)
f1_valid = f1_score(valid_target_binary, valid_preds_binary)
f1_test = f1_score(test_target_binary, test_preds_binary)

print("Linear Regression:")
print("Training Accuracy:", train_accuracy)
print("Validation Accuracy:", valid_accuracy)
print("Testing Accuracy:", test_accuracy)
print("Mean Squared Error:", mse)
print("R2 Score:", r2)
print("Training F1 Score:", f1_train)
print("Validation F1 Score:", f1_valid)
print("Testing F1 Score:", f1_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import f1_score

# Transform principal components into polynomial features
poly = PolynomialFeatures(degree=2)  # adjust degree as needed
features_train_poly = poly.fit_transform(principal_components[train_indices])
features_valid_poly = poly.transform(principal_components[valid_indices])
features_test_poly = poly.transform(principal_components[test_indices])

# Initialize Linear Regression model (with polynomial features)
poly_lr_model = LinearRegression()

# Fit the model to the polynomial features
poly_lr_model.fit(features_train_poly, train_target)

# Predict on polynomial features
train_preds_poly = poly_lr_model.predict(features_train_poly)
valid_preds_poly = poly_lr_model.predict(features_valid_poly)
test_preds_poly = poly_lr_model.predict(features_test_poly)

# Calculate metrics
train_accuracy_poly = poly_lr_model.score(features_train_poly, train_target)
valid_accuracy_poly = poly_lr_model.score(features_valid_poly, valid_target)
test_accuracy_poly = poly_lr_model.score(features_test_poly, test_target)
mse_poly = mean_squared_error(test_target, test_preds_poly)
r2_poly = r2_score(test_target, test_preds_poly)

# Define threshold values for F1 score calculation
threshold = 0.5

# Binarize predictions and target values for F1 score calculation
train_preds_poly_binary = np.where(train_preds_poly >= threshold, 1, 0)
valid_preds_poly_binary = np.where(valid_preds_poly >= threshold, 1, 0)
test_preds_poly_binary = np.where(test_preds_poly >= threshold, 1, 0)
train_target_binary = np.where(train_target >= threshold, 1, 0)
valid_target_binary = np.where(valid_target >= threshold, 1, 0)
test_target_binary = np.where(test_target >= threshold, 1, 0)

# Calculate F1 score
f1_train_poly = f1_score(train_target_binary, train_preds_poly_binary)
f1_valid_poly = f1_score(valid_target_binary, valid_preds_poly_binary)
f1_test_poly = f1_score(test_target_binary, test_preds_poly_binary)

print("Polynomial Regression:")
print("Training Accuracy:", train_accuracy_poly)
print("Validation Accuracy:", valid_accuracy_poly)
print("Testing Accuracy:", test_accuracy_poly)
print("Mean Squared Error:", mse_poly)
print("R2 Score:", r2_poly)
print("Training F1 Score:", f1_train_poly)
print("Validation F1 Score:", f1_valid_poly)
print("Testing F1 Score:", f1_test_poly)
```
